```python
import numpy as np
import matplotlib.pyplot as plt

def generateMetricsDict(lines):
    client = {
        "loss": [],
        "accuracy": [],
        "val_loss": [],
        "val_accuracy": []
    }

    for line in lines:
        metrics = line.split("-")
        for metric in metrics:
            key, value = metric.strip().split(" ")
            client[key[:-1]].append(float(value))
    
    return client

def plot(y1, y2, y1_label, y2_label, image=None):
    x1 = np.linspace(0, 1, len(y1))
    x2 = np.linspace(0, 1, len(y2))

    plt.plot(x1, y1, '-o', label=y1_label)
    plt.plot(x2, y2, '-o', label=y2_label)

    plt.xticks([], [])

    plt.legend()

    if image != None:
        plt.savefig(image)
        
    plt.show()
```

## Command prompt of Client 1

```python
client1_val_accuracy = [0.3319999873638153, 0.8080000281333923, 0.8859999775886536, 0.9100000262260437]
```

```python
y1 = client1_val_accuracy

x1 = np.linspace(0, 1, len(client1_val_accuracy))

plt.plot(x1, y1, '-o', label='Server side validation accuracy')

plt.xticks([], [])
plt.legend()

plt.savefig('client1-credit-card.png')
plt.show()
```
